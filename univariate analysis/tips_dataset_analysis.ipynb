```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Load tips dataset
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

# **a. Countplot**

```python
sns.countplot(x="day", data=tips)
plt.xlabel("Day")
plt.ylabel("Count")
plt.title("Countplot of Days (Tips Dataset)")
plt.show()
```

# b. **PieChart**

```python
plt.pie(tips['sex'].value_counts(), labels=tips['sex'].unique(), autopct='%1.1f%%')
```

# **c**. Histogram

```python
plt.hist(tips['total_bill'])
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.title("Histogram of Total Bill (Tips Dataset)")
plt.show()
```

# d. **Distplot**

```python
sns.distplot(tips['total_bill'])
plt.xlabel("Total Bill ($)")
plt.ylabel("Density")
plt.title("Distribution Plot of Total Bill (Tips Dataset)")
plt.show()
```

# e. **Boxplot**

```python
sns.boxplot(tips['total_bill'])
plt.xlabel("Total Bill ($)")
plt.title("Boxplot of Total Bill (Tips Dataset)")
plt.show()
```
